```python
stroke_dataframe['gender'].replace(['Male', 'Female', 'Other'],[0,1,2], inplace=True)
stroke_dataframe['ever_married'].replace(['Yes', 'No'],[0,1], inplace=True)
stroke_dataframe['work_type'].replace(['Private', 'Self-employed', 'Govt_job', 'children', 'Never_worked'],[0,1,2,3,4], inplace=True)
stroke_dataframe['Residence_type'].replace(['Urban', 'Rural'],[0,1], inplace=True)
stroke_dataframe['smoking_status'].replace(['never smoked', 'formerly smoked', 'smokes', 'Unknown'],[0,1,2,3], inplace=True)
stroke_dataframe.head()
```

```python
stroke_dataframe.dropna(inplace=True)
stroke_dataframe.reset_index(drop=True, inplace=True)
stroke_dataframe.head()
```

```python
(len(stroke_dataframe.index))
#214 null rows removed
```

```python
target = stroke_dataframe["stroke"]
input_columns = stroke_dataframe.loc[:, stroke_dataframe.columns != "stroke"]
import sklearn
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input_columns, target, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
print(y_train),(y_test)
import sklearn
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
k=20
KNN_model = KNN(n_neighbors=k)
```

```python
KNN_model.fit(x_train,y_train)
```

```python
stroke_dataframe.head(10)
```

```python
y_hat = KNN_model.predict(x_test)
```

```python
total_squared_error = (np.sum((y_test - y_hat)**2))
mean_squared_error = total_squared_error/len(y_test)
print(mean_squared_error)
```
